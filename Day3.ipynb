```python
import numpy as np
import matplotlib.pyplot as plt

# paramètres
mu = 0      # moyenne
sigma = 1   # écart type

# échantillon
x = np.linspace(-4, 4, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# visualisation
plt.plot(x, y)
plt.title("Distribution normale (μ=0, σ=1)")
plt.xlabel("Valeurs")
plt.ylabel("Densité de probabilité")
plt.show()
```
